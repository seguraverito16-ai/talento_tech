```python
import pandas as pd
Variables={
    "x":[51,54,57,60,65],
    "y":[67,58,61,63,70]
    }

df=pd.DataFrame(Variables)
mediax=df["x"].mean()
mediay=df["y"].mean()
print(mediax)
print(mediay)
df["xm"]=df["x"]-mediax
df["ym"]=df["y"]-mediay
df["xm2"]=df["xm"]**2
df["ym2"]=df["ym"]**2
df["xy"]=df["xm"]*df["ym"]
print(df)
sumaxy=df["xy"].sum()
sumax2=df["xm2"].sum()
sumay2=df["ym2"].sum()
print(sumaxy)
print(sumax2)
print(sumay2)
correlation=sumaxy/((sumax2*sumay2)**0.5)
print("la correlacion es:",correlation)
```
