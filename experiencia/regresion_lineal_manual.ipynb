```python
import pandas as pd
import numpy as np

Variables={
    "y":[78.5,74.3,104.3,87.6,95.9,109.2,102.7,72.5,93.1,115.9,83.8,113.3,109.4],
    "X1":[7,1,11,11,7,11,3,1,2,21,1,11,10],
    "X2":[26,29,56,31,52,55,71,31,54,47,40,66,68],
    "X3":[6,15,8,8,6,9,17,22,18,4,23,9,8],
    "X4":[60,52,20,47,33,22,6,44,22,26,34,12,12]}

df=pd.DataFrame(Variables)

# Create a copy of the DataFrame to manipulate
X_processed = df.copy()

# Add a column of ones for the intercept term
X_processed.insert(0, 'intercept', 1)

# Drop the 'y' column as requested
X_processed = X_processed.drop(columns=['y'])

# Convert the processed DataFrame to a NumPy matrix
X_matrix = X_processed.values

print("Matrix with a column of ones (without 'y' column):")
print(X_matrix)
```

```python
# Calculate the transpose of the matrix
X_matrix_transpose = X_matrix.T

print("Transpose of the matrix:")
print(X_matrix_transpose)
```

```python
# Extract the 'y' column as a column vector
y_vector = df['y'].values.reshape(-1, 1)

print("Vector 'y':")
print(y_vector)
```

```python
# Multiply the transpose of the matrix by the original matrix
result_matrix_multiplication = np.dot(X_matrix_transpose, X_matrix)

print("Result of X_matrix_transpose * X_matrix:")
print(result_matrix_multiplication)
```

```python
# Calculate the inverse of the result_matrix_multiplication
inverse_matrix = np.linalg.inv(result_matrix_multiplication)

print("Inverse of result_matrix_multiplication:")
print(inverse_matrix)
```

```python
# Calculate betas: (X'X)^-1 * X'y
betas = np.dot(np.dot(inverse_matrix, X_matrix_transpose), y_vector)

print("Beta coefficients:")
print(betas)
```
